```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Real Madrid season-2425.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Referee','Date'],inplace=True)
```

```python
df['Points_Home'] = np.where(df['FTR'] == 'H', 3,
                 np.where(df['FTR'] == 'D', 1, 0))
df['Points_Away'] = np.where(df['FTR'] == 'A', 3,
                 np.where(df['FTR'] == 'D', 1, 0))
```

```python
df.head(20)
```

```python
df.head()
```

```python
print(df.columns)
```

```python
df_home = df.groupby('HomeTeam')['Points_Home'].sum().reset_index()
df_away = df.groupby('AwayTeam')['Points_Away'].sum().reset_index()
```

```python
df_home.rename(columns={'HomeTeam': 'Team'}, inplace=True)
df_away.rename(columns={'AwayTeam': 'Team'}, inplace=True)
```

```python
df1=pd.merge(df_home,df_away,how='outer',on='Team')
```

```python
df1['Points']=df1['Points_Home']+df1['Points_Away']
```

```python
df1.sort_values(by='Points',ascending=False,inplace=True)
```

```python
plt.figure(figsize=(20,10))
ax=sb.barplot(x='Team',y='Points',data=df1)
ax.set(xlabel='Teams',ylabel='Points')
plt.title('La_Liga Team Performances')
plt.show()
```

Inference: Barcelona is the winner of La-Liga 2024-25 season.

```python
correlation_shots_goals = df5[['Total Shots', 'Total Goals']].corr().iloc[0, 1]
correlation_sot_goals = df5[['Total SOT', 'Total Goals']].corr().iloc[0, 1]
```

```python
avg_rm_eff = df5['RM Efficiency'].mean()
avg_opp_eff = 1-df5['Save Efficiency'].mean()

correlation_shots_goals, correlation_sot_goals, avg_rm_eff, avg_opp_eff
```

```python
df = pd.read_csv("Real Madrid season-2425.csv")
rm = df[(df['HomeTeam'] == 'Real Madrid') | (df['AwayTeam'] == 'Real Madrid')].copy()
rm['Date'] = pd.to_datetime(rm['Date'], dayfirst=True)
rm.sort_values('Date', inplace=True)
```

```python
rm['Points'] = np.where(
    rm['HomeTeam'] == 'Real Madrid',
    np.where(rm['FTR'] == 'H', 3, np.where(rm['FTR'] == 'A', 0, 1)),
    np.where(rm['FTR'] == 'A', 3, np.where(rm['FTR'] == 'H', 0, 1))
)
```

```python
rm['Month'] = rm['Date'].dt.month
monthly_performance = rm.groupby('Month')['Points'].mean()
```

```python
rm['Rolling_Points'] = rm['Points'].rolling(window=5, min_periods=1).mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(rm['Date'], rm['Rolling_Points'], label='Rolling Avg (5 Matches)', color='royalblue')
plt.plot(rm['Date'], rm['Points'], label='Match Points', alpha=0.4, linestyle='--', marker='o', markersize=4)
plt.title('Real Madrid Performance Over Time (Rolling Avg of Points)')
plt.xlabel('Date')
plt.ylabel('Points')
plt.grid(True)
plt.legend()
plt.show()
```

Inference: The Performance dipped in the month of February and March.
